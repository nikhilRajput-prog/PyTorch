```python
import time
import torch

size = 2000  # start smaller (10000 may crash memory)

a = torch.rand(size, size)
b = torch.rand(size, size)

# ---------------- CPU ----------------
start = time.time()
torch.matmul(a, b)
end = time.time()

cpu_time = end - start
print(f"CPU time: {cpu_time:.4f} sec")

# ---------------- GPU ----------------
if torch.cuda.is_available():
    c = a.to('cuda')
    d = b.to('cuda')

    # warm-up
    torch.matmul(c, d)
    torch.cuda.synchronize()

    start = time.time()
    torch.matmul(c, d)
    torch.cuda.synchronize()
    end = time.time()

    gpu_time = end - start
    print(f"GPU time: {gpu_time:.4f} sec")

    # speedup
    print(f"Speedup (CPU/GPU): {cpu_time / gpu_time:.2f}x")

else:
    print("GPU not available")
```
