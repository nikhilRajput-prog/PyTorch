```python
class SimpleNeuralNetwork():

  def __init__(self, X):

    self.layer_weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
    self.layer_bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def predict(self, X):
    z = torch.matmul(X, self.layer_weights) + self.layer_bias
    predictions = torch.sigmoid(z)
    return predictions

  def calculate_loss(self, predictions, y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    predictions = torch.clamp(predictions, epsilon, 1 - epsilon)

    # Calculate loss
    current_loss = -(y_train_tensor * torch.log(predictions) + (1 - y_train_tensor) * torch.log(1 - predictions)).mean()
    return current_loss
```

```python
optimizer_lr=0.1
num_epochs=25
```

```python
# create model
neural_network_model = SimpleNeuralNetwork(X_train_tensor)

# define loop
for current_epoch in range(num_epochs):

  # forward pass
  predictions = neural_network_model.predict(X_train_tensor)

  # loss calculate
  current_loss = neural_network_model.calculate_loss(predictions, y_train_tensor)

  # backward pass
  current_loss.backward()

  # parameters update
  with torch.no_grad():
    neural_network_model.layer_weights -= optimizer_lr * neural_network_model.layer_weights.grad
    neural_network_model.layer_bias -= optimizer_lr * neural_network_model.layer_bias.grad

  # zero gradients
  neural_network_model.layer_weights.grad.zero_()
  neural_network_model.layer_bias.grad.zero_()

  # print loss in each epoch
  print(f'Epoch: {current_epoch + 1}, Loss: {current_loss.item()}')
```

```python
# model evaluation
with torch.no_grad():
  predictions = neural_network_model.predict(X_test_tensor)
  predictions = (predictions > 0.9).float()
  accuracy = (predictions == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
